```python
import pandas as pd
pd.options.mode.chained_assignment = None  
```

```python
df = pd.read_excel('OUP_long_MF1564_v1.xls')
df
```

drop rows, only when all are NaN values

```python
df = df.dropna(how='all')
df.shape
```

Forward filling the empty country label cells that were empty due to the pivot table format of the data in Excel

```python
countrycol_filled = df.country.fillna(method='ffill')
df.country = countrycol_filled
```

```python
df.drop(columns=['min_age', 'max_age'], inplace=True)
df
```

Check no NaNs remain

```python
df.isna().sum()
```

```python
df['year'] = df['year'].astype(dtype=int)
```
